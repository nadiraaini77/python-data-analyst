# Isi Missing Value

```python
# isi missing value country
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```python
# isi missing value rating
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
# isi missing value duration
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
```

```python
# isi missing value date_added
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
```

```python
# cek missing value
df.isnull().sum()
```

# Ubah date_added menjadi format tanggal

```python
# Hapus spasi kosong
df['date_added'] = df['date_added'].str.strip()
```

```python
# Ubah ke datetime
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
#tambah kolom baru untuk year_added
df['year_added'] = df['date_added'].dt.year
```

```python
#ganti tipe data year_added ke int
df['year_added'] = df['year_added'].astype(int)
```

```python
df.head()
```

# Mengelompokkan rating

```python
# cek data uniq di rating
df['rating'].unique()
```

```python
# ubah data error menjadi nan
df.loc[df['rating'].str.contains('min', na=False), 'rating'] = None
```

```python
# isi missing value rating
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
def rating_group(rating):
    if rating in ['TV-MA', 'R', 'NC-17']:
        return 'Adult'
    elif rating in ['PG-13', 'PG', 'TV-PG', 'TV-14']:
        return 'Teens'
    elif rating in ['TV-Y7', 'TV-Y', 'TV-Y7-FV']:
        return 'Children'
    else:
        return 'General'
```

```python
# buat kolom rating_group
df['rating_group'] = df['rating'].apply(rating_group)
```

```python
df.head()
```

# Membersihkan kolom country

```python
# hapus spasi dan koma
df['country'] = df['country'].str.strip()
```

```python
# pisahkan negara
df['country'] = df['country'].str.split(', ')
```

```python
# explode negara
df = df.explode('country')
```

```python
# hapus spasi
df['country'] = df['country'].str.strip()
```
